# Step2 – Sort 

Let’s sort the source file. Just remember we have already created a data frame source_file so essentially this is our file.

1.	Sort_values is the function to sort the records.
2.	We can give any column name in – by parameter (basically it’s a Key value in terms of AbInitio)
3.	We can manage ascending or descending order as well.
4.	There are numerous options which we can explore inside this function. (Please try to play with some more parameters)


```python
#Sorting the file with State Key
source_file.sort_values(by='State',ascending=False)
```

# Step3 – Filter By Expression

One of the biggest advantages of having the data as a Pandas Dataframe is that Pandas allows us to slice and dice the data in multiple ways.

Often, you may want to subset a pandas dataframe (file) based on one or more values of a specific column. Essentially, we would like to select rows based on one value or multiple values present in a column

```python
#Lets filter the records based on - State
only_newyork = source_file['State'] == 'New York'
#Remember the above is a boolean variable so it will be just checking in the entire file whether in particular row the state is new york or not?
#To fetch the records only with New York state we will pass the boolean variable into our source data frame
source_file_only_ny = source_file[only_newyork]
#Finally we will VIEW Data to check if we have only New York state. remember - head() will give us first 5 rows. so lets check
source_file_only_ny.head()
```

We can apply any operation on the fields to filter the records just like the same way we do in AbInitio.
-	Isin() method is the equivalent of member function in Abinitio
-	Notnull() to get the NULL records from the entire dataset or a particular column depends upon how you do it.

# Step4 – Reformat

Now we came to the most important Abinitio component. The reformat 😊

Yes, In python also we can do literally all the possible things we do in Abinitio reformat. Though I haven’t tried all 😝

So lets explore some the very basic reformat operations.



Let’s start reformatting the state field . let’s do the upper case of state field.

```python
#New York to Upper Case
source_file.State.str.upper()
```

```python
#To check if the values in the columns is numeric or not? 
source_file.State.str.isnumeric()
#We should get the false values for all the columns because we have all the values in State are strings
```

Now lets create an additional column-
-	We have 5 columns currently.
-	We will create an additional columns named – ‘Total Spend’.
-	We will combine the first 3 columns to create this column.



```python
#Creating an additonal columns by combining the first 3 columns 
source_file["Total Spend"] = source_file["R&D Spend"] + source_file["Administration"] + source_file["Marketing Spend"]
```

```python
#Checking if the columns creating correctly or not?
source_file.shape
```

You can see now we have 1000 rows and 6 columns (One additional we just created). This additonal column is the SUM of the first 3 columns.

```python
#Lets VIEW the first 5 records to check how the data is looking currently.
source_file.head()
```

WOW! we have our newly created column :)
